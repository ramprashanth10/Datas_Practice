```python
y = ipl['player_of_match'].value_counts()[0:5]
```

```python
plt.figure(figsize=(8,8))
plt.bar(x,y, color="g")
plt.show()
```

```python
batting_first = ipl[ipl['win_by_runs'] != 0]
```

```python
batting_first['win_by_runs'].max()
```

```python
plt.figure(figsize = (8,8))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.show()
```

```python
y = batting_first['winner'].value_counts()
```

```python
x = batting_first['winner'].value_counts().keys()
```

```python
x
```

```python
plt.figure(figsize=(6,6))
plt.bar(x,y, color = ['blue','yellow','red'])
plt.title('Most winners batting first')
plt.xlabel('Teams')
plt.ylabel('Runs')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(y,labels = x, autopct = '%0.1f%%')
plt.show()
    
```

```python
batting_second = ipl[ipl['win_by_wickets'] != 0]
```

```python
y = batting_second['winner'].value_counts()
```

```python
x = batting_second['winner'].value_counts().keys()
```

```python
plt.figure(figsize=(6,6))
plt.bar(x,y,color = ['purple','blue', 'red'])
plt.title('Batting second won')
plt.xlabel('Teams')
plt.ylabel('no.wins')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.pie(y, labels = x, autopct = '%0.1f%%')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.hist(batting_second['win_by_wickets'], bins = 30)
plt.show()
```
